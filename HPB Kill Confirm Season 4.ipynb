```python
#Combo Calculators
#Calculates damage of given combo, returns combo damage and remaining HP as a list.
def comboCalculator(comboList,startingRISC,startingHP,index,counterHit=0,PoboP1=False,PoboP2=False,charModifier=0):#Minimum arguments are combo, starting RISC, starting HP, character index
    damage = 0
    currentRISC = startingRISC
    currentHP = startingHP
    currentProration = 100
    if(comboList[0]==StandD):
        comboList[0]=StandDStarter
    if(comboList[0]==StandChargedD):
        comboList[0]=StandChargedDStarter  
    for move in comboList:  
        if(verbose or summary):
            print(move["Name"])
        damage = damage + moveDamage(move["Damage"],(Defense[index]),currentGuts(currentHP,Guts[index]),(RISCFactor(currentRISC)),currentProration,move["Minscaling"],index,counterHit,PoboP1,PoboP2,charModifier)
        counterHit = 0
        if(currentRISC>=0 and move!=PotBuster and move!=Throw and move!=Heavenly and move!=HeatKnuckleHit and move!=HeatExtend):
            currentRISC=currentRISC*75
            currentRISC=math.trunc(currentRISC/100)
        currentRISC = currentRISC - move["RISC Loss"]
        if(comboList.index(move)==0):
            currentProration = comboList[0]["Proration"]
        if(move==PotBuster or move==Throw):
            if(currentProration>move["Proration"]):
              currentProration = move["Proration"]
        currentHP = startingHP - damage      
        if(move==HeatKnuckle):
            i=0
            while(i<8):
                damage = damage + moveDamage(HeatKnuckleHit["Damage"],(Defense[index]),currentGuts(currentHP,Guts[index]),(RISCFactor(currentRISC)),currentProration,move["Minscaling"],PoboP1,PoboP2)
                if(verbose):
                    print(damage)
                currentHP = startingHP - damage      
                i=i+1
            if(verbose):   
                print(HeatExtend["Name"])    
            damage = damage + moveDamage(HeatExtend["Damage"],(Defense[index]),currentGuts(currentHP,Guts[index]),(RISCFactor(currentRISC)),currentProration,move["Minscaling"],PoboP1,PoboP2)
            if(verbose):
                print(damage)
            currentHP = startingHP - damage
            if(currentProration>HeatExtend["Proration"]):
              currentProration = HeatExtend["Proration"]
            currentRISC = currentRISC - HeatExtend["RISC Loss"]  
        if(verbose or summary):   
            print("Current Combo") 
            print(damage)
            print(currentRISC)
            print(currentHP)
    return [damage, currentHP]      

#Finds maximum HP at which a given combo will always kill. Returns 2 lists - WIK that stores 1 or 2 depending whether the combo kills or not, and MinWIK which stores the minimum HP threshold that gurantees a kill
def killThreshold(combo,startingRISC,counterHit=False,PoboP1=False,PoboP2=False,):#Minimum arguments are combo and starting RISC
    WIK = np.zeros(shape=(len(Character), 420))
    MinWIK = np.zeros(shape=(len(Character)))
    minHP = np.zeros(420)
    for x in Character:
        index = Character.index(x)
        startingHP = 1
        currentHP = 1
        flag = True
        while(startingHP<=420):
            currentHP=startingHP
            minHP[startingHP-1]=currentHP
            currentHP = comboCalculator(combo,startingRISC,currentHP,index,counterHit,PoboP1,PoboP2)[1]
            if currentHP<=0:
                WIK[index][startingHP-1]=2
                if(flag):
                    MinWIK[index]=startingHP
            else:
                WIK[index][startingHP-1]=1
                flag = False
            startingHP=startingHP+1
    return [WIK,MinWIK]

#Finds highest HP at which a given combo kills. Returns the same lists as killThreshold, but ignores Guts thresholds that can cause a combo to not kill.
def killUpper(combo,startingRISC,counterHit=False,PoboP1=False,PoboP2=False):#Minimum arguments are combo and starting RISC
    WIK = np.zeros(shape=(len(Character), 420))
    MinWIK = np.zeros(shape=(len(Character)))
    minHP = np.zeros(420)
    for x in Character:
        index = Character.index(x)
        startingHP = 1
        currentHP = 1
        flag = True
        while(startingHP<=420):
            currentHP=startingHP
            minHP[startingHP-1]=currentHP
            currentHP = comboCalculator(combo,startingRISC,currentHP,index,counterHit,PoboP1,PoboP2)[1]
            if currentHP<=0:
                WIK[index][startingHP-1]=2
                MinWIK[index]=startingHP
            startingHP=startingHP+1
    return [WIK,MinWIK]
```
